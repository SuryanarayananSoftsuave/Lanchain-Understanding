### Core difference between invoke() and format() in PromptTemplate

invoke()

- Executes the PromptTemplate
- Returns a LangChain PromptValue
- Used as input to an LLM via llm.invoke(prompt_value)
- Not pipeable with |

format()

- Executes the PromptTemplate
- Returns a plain Python str
- Used when you want to manually pass text to something
- Not pipeable with |

```python
from langchain_core.prompts import PromptTemplate

# PromptTemplate
template = PromptTemplate(
    template="Tell a joke about, {name}, in a {style} style which must be in the {length} words",
    input_variables=["name", "style", "length"],
    validate_template=True
)

# invoke() returns a LangChain PromptValue object (for chaining), can be used for further processing in the chain
prompt = template.invoke({
    "name": "John",
    "style": "dad",
    "length": "100 words"
})
print(prompt)

# format() returns a plain Python string (for direct use), it is used when we want to use the prompt directly without any further processing in the chain
text = template.format(
    name="jim",
    style="laughter",
    length="100 words"
)
print(text)

print(type(template))  # <class 'langchain_core.prompts.prompt.PromptTemplate'>
print(type(prompt))   # <class 'langchain_core.prompts.prompt.PromptValue'>
print(type(llm)) # <class 'langchain_openai.chat_models.base.ChatOpenAI'>
print(type(llm.invoke(input=prompt))) # <class 'langchain_core.prompts.prompt.PromptValue'>


print(isinstance(template, Runnable)) # True 
print(isinstance(prompt, Runnable)) # False so it cant be used in a chain 
print(isinstance(llm, Runnable)) # True  
```

We can also use the F sting but in the prompt template we can validate and save the prompt in a file as text or yaml

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    template="Tell a joke about, {name}, in a {style} style which must be in the {length} words",
    input_variables=["name", "style", "length"],
    validate_template=True
)

template.save("prompt_template.yaml")
```

```python
from langchain_core.prompts import load_prompt

template = load_prompt("prompt_template.yaml")


prompt = template.invoke({
    "name": "John",
    "style": "dad",
    "length": "100 words"
})
print(prompt)

text = template.format(
    name="jim",
    style="laughter",
    length="100 words"
)
print(text)
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

print(prompt)
response = llm.invoke(prompt)

print(response.content)
```
