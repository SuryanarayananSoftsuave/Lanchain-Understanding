## Runnables


The Below code run(), LLMChain some of then are deprecated but still we can use using Langchain-classic , but not recommended. It will provide the Deprecated Warning for those things.

```python
from langchain_classic.chains import LLMChain  
from langchain_openai import OpenAI            
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
import os

load_dotenv()

llm = OpenAI(
    model_name="gpt-4o-mini", 
    temperature=0.7
)

prompt = PromptTemplate(
    input_variables=["topic"],
    template="Generate a new summary about the topic: {topic}"
)

# Using the legacy LLMChain class from langchain-classic
chain = LLMChain(llm=llm, prompt=prompt)

topic = input("Enter Topic: ")
response = chain.run({"topic": topic})

print(response)
```
